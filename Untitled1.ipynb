```python
import speech_recognition as sr
import pyttsx3 as p
from fuzzywuzzy import process
import time

# Initialize the text-to-speech engine
engine = p.init()
rate = engine.getProperty('rate')
engine.setProperty('rate', 160)
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)

# Function to speak text
def speak(text):
    engine.say(text)
    engine.runAndWait()

# Define a dictionary of questions and answers
qa_pairs = {
    "how can i update my profile details": "You can update your profile details by going to the Profile Management section under Settings.",
    "how do i add my recent achievements or collaborations": "To add your recent achievements or collaborations, navigate to the 'Achievements' section in your profile and click 'Add New'.",
    "can i hide certain information on my profile": "Yes, you can hide certain information on your profile by adjusting your privacy settings in the Profile Management section.",
    "what brands are currently looking for collaborations": "To find brands looking for collaborations, check the 'Collaboration Opportunities' section under the Explore tab.",
    "how do i set preferences for the types of brands i want to work with": "You can set your brand preferences by going to your profile settings and selecting 'Collaboration Preferences'.",
    "can you show me the top campaigns in my niche": "Sure! You can view the top campaigns in your niche by visiting the 'Campaigns' section and filtering by your interests.",
    # Add other entries here...
}

# Function to match the user's query to the closest question using fuzzy matching
def get_answer(query):
    query = query.lower()
    # Find the best match from the dictionary keys
    best_match, score = process.extractOne(query, qa_pairs.keys())
    # Set a threshold score for matching
    if score > 70:  # You can adjust this threshold as needed
        return qa_pairs[best_match]
    else:
        return "Sorry, I don't have an answer for that."

# Function to listen for user input with a timeout
def listen_for_question(timeout=10):
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        recognizer.energy_threshold = 10000
        recognizer.adjust_for_ambient_noise(source, 1.2)
        print("Listening...")
        try:
            audio = recognizer.listen(source, timeout=timeout)
            user_query = recognizer.recognize_google(audio)
            print(f"You said: {user_query}")
            return user_query.lower()
        except sr.UnknownValueError:
            speak("Sorry, I didn't catch that. Could you please repeat?")
        except sr.RequestError:
            speak("Sorry, my speech service is down. Please try again later.")
        except Exception as e:
            print("No input received within the time limit.")
            speak("I didn't hear anything. Please ask your question.")
        return None

# Start the conversation
speak("Hello sir, I am your Buzz assistant. How are you?")
text = listen_for_question(timeout=10)

if text and "what about you" in text:
    speak("I am also having a good day, ma'am.")
speak("What can I do for you?")

# Loop to allow multiple questions with a time limit
while True:
    user_query = listen_for_question(timeout=10)
    if user_query:
        # Fetch the answer based on the user query
        answer = get_answer(user_query)
        print(f"Answer: {answer}")
        speak(answer)
        
        # Wait before listening for the next question
        time.sleep(2)  # Wait for 2 seconds before allowing the next question

        # Ask if the user wants to ask another question
        speak("Do you want to ask another question? Say 'yes' to continue or 'no' to exit.")
        response = listen_for_question(timeout=5)
        if response and "no" in response:
            speak(" Ok Goodbye! Have a great day!")
            break
        elif response and "yes" in response:
            speak("Okay, go ahead with your next question.")
        else:
            speak("I didn't catch that. Please ask your next question or say 'no' to exit.")
    else:
        # If no input was received, ask the user to ask the question again
        speak("I didn't hear anything. Please ask your question.")
```
